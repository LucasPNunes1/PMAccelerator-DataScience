## 9. Conclusion & Architecture Decision

**(Note on Z-Score Scaling):** We did *not* apply standard scaling (Z-score) to the dataset. For the models chosen in our robust ensemble architecture, feature scaling is unnecessary, thus optimizing the pipeline's computational efficiency.

By building deeply analytical feature sets without over-normalization, we identified that a **Diverse Ensemble Pipeline (Gradient Boosting + Linear Regression + Prophet)** acts as the definitive best-case production model. 

**Why is it the best?**
- **Linear Regression** acts as our stable anchor. Because the strongest signals are `rolling_mean_7` and `lag_1`, LR perfectly captures the immediate autoregressive trend without overfitting.
- **Gradient Boosting** provides non-linear nuance, capturing sharp, complex thresholds (like sudden gusts or extreme weather interactions) that a simple line cannot.
- **Prophet** guarantees long-term seasonal accuracy, supplying a pure mathematical baseline for the annual cycle.

Averaging these three entirely different mathematical architectures cancels out individual weaknesses (like tree overfitting), yielding the ultimate robust and stable forecast for future weather trends.
